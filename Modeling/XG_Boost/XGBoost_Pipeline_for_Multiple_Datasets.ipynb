```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import (confusion_matrix, classification_report,
                             f1_score, accuracy_score, recall_score)
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# --- Global Configuration ---

# Columns to drop before training (likely common across all datasets)
DROP_COLS = ['start_time', 'end_time', 'cow']

# Base XGBoost Hyperparameters (num_class will be set dynamically)
BASE_XGB_CONFIG = {
    'objective': 'multi:softmax',
    'eval_metric': 'mlogloss',
    'random_state': 42,
    'learning_rate': 0.2,
    'max_depth': 7,
    'n_estimators': 200,
    'subsample': 0.9
}

def load_data(path: str, dataset_name: str):
    """Safely loads a CSV file."""
    try:
        df = pd.read_csv(path)
        print(f"   -> Loaded {dataset_name} data from: {path} ({len(df)} rows)")
        return df
    except FileNotFoundError:
        print(f"ðŸ›‘ Error: {dataset_name} file not found at path: {path}")
        return None
    except Exception as e:
        print(f"ðŸ›‘ Error loading {dataset_name} dataset from {path}: {e}")
        return None

def evaluate_model_and_plot(model, X, y_true, le: LabelEncoder, analysis_name: str, split_name: str):
    """
    Evaluates the trained model on a given dataset split, prints metrics,
    and visualizes the confusion matrix.
    """
    print(f"\n--- Evaluation for {analysis_name} ({split_name}) ---")

    # 1. Prediction
    try:
        y_pred = model.predict(X)
    except xgb.core.XGBoostError as e:
        print(f"Error during prediction on {split_name} for {analysis_name}: {e}")
        return

    # 2. Global metrics
    # Note: Handles case where y_true might only contain one class for smaller splits
    if len(np.unique(y_true)) > 1:
        f1 = f1_score(y_true, y_pred, average='macro')
        recall = recall_score(y_true, y_pred, average='macro')
    else:
        # If only one class is present, macro average metrics are not well-defined
        # Using weighted average for robustness with zero_division=0
        f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)
        recall = recall_score(y_true, y_pred, average='weighted', zero_division=0)
        
    accuracy = accuracy_score(y_true, y_pred)

    print("ðŸŽ¯ Global Metrics")
    # Indicate that macro/weighted is used for clarity
    print(f"F1 Score (Macro/Weighted*): {f1:.4f}") 
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Recall (Macro/Weighted*): {recall:.4f}\n")

    # 3. Classification report
    print("ðŸ“‹ Classification Report (Per Class):")
    # Setting zero_division to 0 ensures clean output even if a class is entirely missing from a split
    # We now explicitly pass the labels (0, 1, 2, ... N-1) to match the length of target_names (le.classes_)
    print(classification_report(
        y_true, 
        y_pred, 
        target_names=le.classes_, 
        labels=np.arange(len(le.classes_)), # Explicitly use all encoded labels
        zero_division=0
    ))

    # 4. Confusion Matrix Calculation
    cm = confusion_matrix(y_true, y_pred)
    # Handle division by zero if a row sum is zero (i.e., class missing from true labels)
    row_sums = cm.sum(axis=1)
    # np.divide ensures safe division by zero, resulting in 0 where division is impossible
    cm_normalized = np.divide(cm.astype('float'), row_sums[:, np.newaxis], out=np.zeros_like(cm, dtype=float), where=row_sums[:, np.newaxis]!=0)

    # 5. Plotting
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm_normalized, annot=False, cmap='Blues', cbar=True)

    # Add counts and normalized percentages to the heatmap
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            # Display count and percentage in the center of the cell
            plt.text(j+0.5, i+0.5, f"{cm[i, j]}\n({cm_normalized[i,j]*100:.2f}%)",
                     ha='center', va='center',
                     fontsize=10,
                     color='white' if cm_normalized[i,j] > 0.5 else 'black')

    # Tick labels
    plt.xticks(np.arange(len(le.classes_)) + 0.5, le.classes_, rotation=45, ha='right')
    plt.yticks(np.arange(len(le.classes_)) + 0.5, le.classes_, rotation=0)
    plt.title(f'Confusion Matrix for {analysis_name} ({split_name})', fontsize=14, weight='bold')
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.tight_layout()
    plt.show()


def run_xgboost_pipeline(analysis_name: str, main_data_path: str, unseen_path: str, drop_cols: list, target_cols: list, base_xgb_config: dict, test_size: float = 0.2):
    """
    Runs the full XGBoost training and evaluation pipeline.
    It splits the data from `main_data_path` into Training and Testing sets,
    and uses `unseen_path` for final evaluation.
    
    The number of classes for XGBoost is determined dynamically by the length of target_cols.
    """
    print(f"\n=======================================================")
    print(f"ðŸš€ Starting Pipeline for: {analysis_name}")
    print(f"=======================================================")

    # 1. Load the two primary datasets
    df_main = load_data(main_data_path, f"{analysis_name} Main (Train/Test)")
    df_unseen = load_data(unseen_path, f"{analysis_name} Unseen")

    if df_main is None or df_unseen is None:
        print(f"ðŸ›‘ Pipeline aborted for {analysis_name} due to missing data.")
        return

    # 1.5 Split the main data into Training and Testing sets
    print(f"   -> Splitting main data (Train/Test) with test_size={test_size}")
    df_train, df_test = train_test_split(df_main, test_size=test_size, random_state=42)
    
    # 2. Drop unnecessary columns from all splits
    df_train.drop(columns=drop_cols, errors='ignore', inplace=True)
    df_test.drop(columns=drop_cols, errors='ignore', inplace=True)
    df_unseen.drop(columns=drop_cols, errors='ignore', inplace=True)
    
    # 3. Check if target columns exist after dropping
    missing_target_cols = [col for col in target_cols if col not in df_train.columns]
    if missing_target_cols:
        print(f"ðŸ›‘ Error: Target columns {missing_target_cols} not found in the primary (Training) dataset's feature space.")
        return

    # 4. Multi-class target creation
    # The max value across the target columns determines the class (e.g., if 'calving' is 1 and all others are 0, the target is 'calving')
    try:
        df_train['target'] = df_train[target_cols].idxmax(axis=1)
        df_test['target'] = df_test[target_cols].idxmax(axis=1)
        df_unseen['target'] = df_unseen[target_cols].idxmax(axis=1)
    except KeyError as e:
        print(f"ðŸ›‘ Error: One of the specified target columns is missing in a split for {analysis_name}: {e}")
        return

    # 5. Label encoding: Define classes explicitly and then transform
    le = LabelEncoder()
    # FIT the encoder using the specific target column names (the strings),
    # ensuring the internal class list is exactly the length of target_cols.
    le.fit(target_cols) 
    
    # Transform data using the fitted encoder
    try:
        y_train = le.transform(df_train['target'])
        y_test = le.transform(df_test['target'])
        y_unseen = le.transform(df_unseen['target'])
    except ValueError as e:
        # This will now catch cases where a label exists in the data (e.g., due to null values or data contamination)
        # that was not in the expected target_cols list.
        print(f"ðŸ›‘ Error: Data contains classes not defined in 'target_cols' (e.g., unexpected values or NaNs). Ensure the values in the 'target' column strictly match the strings in 'target_cols'. Details: {e}")
        return
        
    print(f"   -> Classes found and encoded (based on target_cols): {le.classes_}")

    # 6. Split features and labels
    X_train = df_train.drop(columns=target_cols + ['target'])
    X_test = df_test.drop(columns=target_cols + ['target'])
    X_unseen = df_unseen.drop(columns=target_cols + ['target'])
    
    # Final check to ensure feature sets are consistent
    if not X_train.columns.equals(X_test.columns) or not X_train.columns.equals(X_unseen.columns):
        print(f"âš ï¸ Warning: Feature columns are inconsistent between splits for {analysis_name}. Attempting column alignment.")
        # Attempt to align columns to prevent prediction errors, filling missing features with 0
        common_cols = X_train.columns.tolist()
        X_test = X_test.reindex(columns=common_cols, fill_value=0)
        X_unseen = X_unseen.reindex(columns=common_cols, fill_value=0)
    
    print(f"   -> Training samples: {len(X_train)}, Testing samples: {len(X_test)}, Unseen samples: {len(X_unseen)}")

    # 7. XGBoost model initialization and training
    # Prepare parameters, setting num_class dynamically based on the current dataset's labels
    xgb_params = base_xgb_config.copy()
    xgb_params['num_class'] = len(target_cols)
    
    model = xgb.XGBClassifier(**xgb_params)
    model.fit(X_train, y_train)
    print("   -> Model trained successfully.")

    # 8. Evaluation on Test Data
    evaluate_model_and_plot(model, X_test, y_test, le, analysis_name, "Test Set (20% Split)")

    # 9. Evaluation on Unseen Data
    evaluate_model_and_plot(model, X_unseen, y_unseen, le, analysis_name, "Unseen Set (Separate File)")
    
    print(f"\nâœ… Pipeline finished for: {analysis_name}")
    print("-------------------------------------------------------")
    
    return model

# --- Main Execution Loop ---

# Define the paths and LABELS for your four datasets.
# IMPORTANT: You MUST update the 'main_data_path', 'unseen_path', 
# and the 'target_cols' for each dataset configuration.
datasets_to_run = [
 {
        'name': 'Dataset 1 Analysis ',
        'target_cols': ['oestrus', 'mastitis', 'calving', 'lameness', 'other_disease', 'ok'],
        'main_data_path': r"C:/Users/lamia/Downloads/dataset1_balanced_trainval.csv",
        'unseen_path': r"C:/Users/lamia/Downloads/dataset1_balanced_holdout.csv",
        'test_size': 0.2
    },
    {
        'name': 'Dataset 2 Analysis',

        'target_cols': ['oestrus', 'mastitis', 'lameness', 'other_disease', 'ok'],
        'main_data_path': r"C:/Users/lamia/Downloads/dataset2_balanced_trainval.csv",
        'unseen_path': r"C:/Users/lamia/Downloads/dataset2_balanced_holdout.csv",
        'test_size': 0.2
    },
    {
        'name': 'Dataset 3 Analysis',

        'target_cols': ['oestrus', 'ok'],
        'main_data_path': r"C:/Users/lamia/Downloads/dataset3_balanced_trainval.csv",
        'unseen_path': r"C:/Users/lamia/Downloads/dataset3_balanced_holdout.csv",
        'test_size': 0.2
    },

    {
        'name': 'Dataset 4 Analysis',

        'target_cols': ['oestrus', 'calving','mastitis', 'lameness', 'other_disease', 'ok'],
        'main_data_path': r"C:/Users/lamia/Downloads/dataset4_balanced_trainval.csv",
        'unseen_path': r"C:/Users/lamia/Downloads/dataset4_balanced_holdout.csv",
        'test_size': 0.2
    }
]

# Run the pipeline for all defined datasets
trained_models = {}
for config in datasets_to_run:
    # Use .get() for test_size to allow default value (0.2) if not specified
    model = run_xgboost_pipeline(
        analysis_name=config['name'],
        main_data_path=config['main_data_path'],
        unseen_path=config['unseen_path'],
        drop_cols=DROP_COLS,
        target_cols=config['target_cols'],
        base_xgb_config=BASE_XGB_CONFIG,
        test_size=config.get('test_size', 0.2)
    )
    if model:
        # Store the trained model if successful
        trained_models[config['name']] = model

print("\n\nAll analysis pipelines completed.")
print(f"Successfully trained and evaluated models for: {list(trained_models.keys())}")
```
